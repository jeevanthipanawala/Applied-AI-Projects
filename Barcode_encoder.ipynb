```python
from PIL import Image, ImageDraw # Python Image libraries were imported to draw the bars

def get_bar_width(letter):
    # The letter should be converetd to lower case as the encoder does not take the case into account
    # Get the position of it in the alphabet
    # To get the position we have to first get the unicode value of the letter and substract charater a's unicode value from it. 
    # As this gives 0 to a, add 1 to calculation. Then it correctly give the position number of the letter in the alphabet
    position=ord(letter.lower())-ord('a')+1
    bar_width=position+1 # Width=position+1
    return bar_width

# Create Barcode
def encode_string(name):
    # Draw canvas
    image = Image.new("RGB", (800, 400), "white")
    draw = ImageDraw.Draw(image)
    # Variables
    step=9
    x=step-1 # Start with 9th pixel
    # Draw bars for the string given. Loop through the characters
    for char in name:
        if char==' ': 
            bar_width=1
            draw.rectangle([(x,150),(x+bar_width,250)],fill="black") # spaces represented by a bar of width=1,height=(250-150)
        else:
            bar_width=get_bar_width(char)
            draw.rectangle([(x,10),(x+bar_width,350)],fill="black") # other character width is generated by get_bar_width function. height=(350-10)
        x+=bar_width+step # next starting x cordinate= current x + bar_width(1 for spaces)+step
    image.save("Output.png")
```

```python
encode_string("Abbas Cheddad")
```
